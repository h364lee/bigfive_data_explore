## Plot trait distributions

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style='whitegrid')
```

```python
#Random sampling for plotting to reduce the number of rows for visual clarity.
df_sample = df.sample(n=1000, random_state=12)
```

#### Individual Histograms for Traits

```python
#Openness to Experience Distribution Histogram
opn_mean = df_sample['Openness'].mean()

plt.figure(figsize=(8, 5))
sns.histplot(df_sample['Openness'], bins=30, kde=True, color='skyblue')
plt.axvline(opn_mean, color='red', linestyle='--', label=f'Mean: {opn_mean:.2f}')
plt.title('Openness Distribution')
plt.xlabel('Openness Score')
plt.ylabel('Frequency')

plt.legend()

plt.show()
```

```python
#Conscientiousness Distribution Histogram
csn_mean = df_sample['Conscientiousness'].mean()

plt.figure(figsize=(8, 5))
sns.histplot(df_sample['Conscientiousness'], bins=30, kde=True, color='skyblue')
plt.axvline(csn_mean, color='red', linestyle='--', label=f'Mean: {csn_mean:.2f}' )
plt.title('Conscientiousness Distribution')
plt.xlabel('Conscientiousness Score')
plt.ylabel('Frequency')

plt.legend()
plt.show()
```

```python
#Extraversion Distribution Histogram
ext_mean = df_sample['Extraversion'].mean()

plt.figure(figsize=(8, 5))
sns.histplot(df_sample['Extraversion'], bins=30, kde=True, color='skyblue')
plt.axvline(ext_mean, color='red', linestyle='--', label=f'Mean: {ext_mean:.2f}' )
plt.title('Extraversion Distribution')
plt.xlabel('Extraversion Score')
plt.ylabel('Frequency')

plt.legend()

plt.show()
```

```python
#Agreeableness Distribution Histogram
agr_mean = df_sample['Agreeableness'].mean()

plt.figure(figsize=(8, 5))
sns.histplot(df_sample['Agreeableness'], kde=True, color='skyblue', bins=30)
plt.axvline(agr_mean, color='red', linestyle='--', label=f'Mean: {agr_mean:.2f}')
plt.title('Agreeableness Distribution')
plt.xlabel('Agreeableness Score')
plt.ylabel('Frequency')

plt.legend()
plt.show()
```

```python
#Neuroticism Distribution Histogram
nrt_mean = df_sample['Neuroticism'].mean()

plt.figure(figsize=(8, 5))
sns.histplot(df_sample['Neuroticism'], bins=15, kde=True, color='skyblue')
plt.axvline(nrt_mean, color='red', linestyle='--', label=f'Mean: {nrt_mean:.2f}')
plt.title('Neuroticism Distribution')
plt.xlabel('Neuroticism Score')
plt.ylabel('Frequency')

plt.legend()
plt.show()
```

#### 5 Histograms in a row (subplots)

```python
traits = ['Openness', 'Conscientiousness', 'Extraversion', 'Agreeableness', 'Neuroticism']

means = [df_sample[trait].mean() for trait in traits]

fig, axes = plt.subplots(1, 5, figsize=(25, 5), sharey=True)

for i, trait in enumerate(traits): #enumerate: loop over a list while keeping track of the index (Automatically gives you a (index, value) pair on each loop.)
    ax = axes[i]
    sns.histplot(df_sample[trait], bins=30, kde=True, color='skyblue', ax=ax)
    ax.axvline(means[i], color='red', linestyle='--', label=f'Mean: {means[i]:.2f}')
    ax.set_title(f'{trait} Distribution')
    ax.set_xlabel(f'{trait} Score')
    ax.set_xlim(1, 5) #standardizes x-scales
    ax.set_ylabel('Frequency' if i == 0 else '')
    ax.legend()
plt.tight_layout()
plt.show()
```

#### 5 Histograms in two rows (subplots)

```python
traits = ['Openness', 'Conscientiousness', 'Extraversion', 'Agreeableness', 'Neuroticism']
means = [df_sample[trait].mean() for trait in traits]

# 2 rows x 3 columns
fig, axes = plt.subplots(2, 3, figsize=(18, 8), sharey=True)

# Flatten axes for easy looping
axes_flat = axes.flatten()

for i, trait in enumerate(traits):
    ax = axes_flat[i]
    sns.histplot(df_sample[trait], bins=30, kde=True, color='skyblue', ax=ax)
    ax.axvline(means[i], color='red', linestyle='--', label=f'Mean: {means[i]:.2f}')
    ax.set_title(f'{trait} Distribution')
    ax.set_xlabel(f'{trait} Score')
    ax.set_xlim(1, 5)
    if i % 3 == 0:  # Only leftmost plots show y-axis label
        ax.set_ylabel('Frequency')
    else:
        ax.set_ylabel('')
    ax.legend()

# Hide unused 6th subplot (bottom right)
axes_flat[-1].set_visible(False)

plt.tight_layout()
plt.show()
```
